```python
import requests
from bs4 import BeautifulSoup
```

```python
with open("./musinsa.text","w",encoding="utf8") as f:

    for i in range(1,11):
        url="https://www.musinsa.com/ranking/best?period=now&age=ALL&mainCategory=003&subCategory=&leafCategory=&price=&golf=false&kids=false&newProduct=true&exclusive=false&discount=false&soldOut=false&page={}&viewType=small&priceMin=&priceMax=".format(i)
        headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36"}
        res=requests.get(url,headers=headers)
        res.raise_for_status()

        soup=BeautifulSoup(res.text,"lxml")

        items=soup.find_all("div",attrs={"class":"article_info"})

        for item in items:
            brand=item.find("p",attrs={"class":"item_title"}).get_text().strip()

            product=item.find("p",attrs={"class":"list_info"})
            if product.find('strong'):
                product.find('strong').decompose()
            product=product.get_text().strip()
            
        
            price=item.find("p",attrs={"class":"price"})
            if price.find('del'):
                price.find('del').decompose()
            price=price.get_text().strip()
            
            f.write(brand+"|"+product+"|"+price+"\n") 
```

```python
class TrieNode:
    def __init__(self, key, data=None):
        self.key = key
        self.data = []
        self.children = {}
        
head = TrieNode(None)
```

```python
for bName in bName_list:
    cur = head
    for b in bName:
        if b not in cur.children:
            cur.children[b] = TrieNode(b)
            cur.children[b].data.append(bName)
        else:
            cur.children[b].data.append(bName)
        cur = cur.children[b]
```

```python
from jamo import h2j, j2hcj, j2h

text = "프라다"
return_value = ""
for t in text:
    jamo_list = j2hcj(h2j(t))
    if len(jamo_list) == 2:
        return_value += jamo_list[0]
        return_value += j2h(*jamo_list)
    else:
        return_value += jamo_list[0]
        for i in range(2,len(jamo_list)+1):
            han = j2h(*jamo_list[:i])
            return_value += han
```
